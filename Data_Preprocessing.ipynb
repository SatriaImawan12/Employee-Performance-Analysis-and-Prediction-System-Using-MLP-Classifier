#**5. Check Outliers and Impute Outliers**

**BoxPlot:** Box plots digunakan untuk menunjukkan distribusi nilai data numerik serta membantu mendeteksi outlier (nilai ekstrem).

```python
out = data[['Age','DistanceFromHome','EmpHourlyRate','EmpLastSalaryHikePercent',
                   'TotalWorkExperienceInYears','TrainingTimesLastYear','ExperienceYearsAtThisCompany',
                   'ExperienceYearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']]

plt.figure(figsize=(20,18)) # defining canvas size
plotno = 1 # counter

for column in out: # iteration of columns / acessing the columns from count
    if plotno<=12:    # set the limit
        plt.subplot(4,3,plotno) # # plotting 7 graphs (3-rows,3-columns) ,plotnumber is for count
        sns.boxplot(out[column]) # Plotting box plots to detect the outlier
        plt.xlabel(column,fontsize=20)  # assigning name to x-axis and font size is 20
    plotno+=1 # counter increment
plt.tight_layout()
plt.show() # used to hide the storage loction
```

* Berdasarkan boxplot diatas, kita bisa melihat ada nilai outliers dibeberapa kolom.

* Selanjutnya, kita akan melakukan imputasi outliers menggunakan median, karena berdasarkan EDA yang sudah kita lakukan, distribusi data dibeberapa kolom ada yang berupa normal skewness, positive skewness, dan negative skewness. Oleh karena itu, median menjadi nilai paling cocok untuk imputasi.

* Disini kita akan melakukan pengecekan outliers berdasarkan IQR (Q3 - Q1)

##**5.1 Total Work Experience in Year**

```python
# Use iqr range because some skewed in data
iqr = stats.iqr(data['TotalWorkExperienceInYears'],interpolation='midpoint')
print("IQR:",iqr)

# Getting first & third quantile
Q1 = data['TotalWorkExperienceInYears'].quantile(0.25)
Q3 = data['TotalWorkExperienceInYears'].quantile(0.75)

# Get maximum and minimum limit
min_limit = Q1 - 1.5*iqr
print("Minimum limit:",min_limit)

max_limit = Q3 + 1.5*iqr
print("Maximum Limit:",max_limit)
```

```python
# Impute outlier
data.loc[data['TotalWorkExperienceInYears'] > max_limit,'TotalWorkExperienceInYears']=data['TotalWorkExperienceInYears'].median()
```

```python
# filtering data after imputation
data.loc[data['TotalWorkExperienceInYears'] > max_limit]
```

#**7. Feature Engineering**

* Berdasarkan EDA (Exploratory Data Analysis) yang dilakukan, kita mendapati bahwa fitur "tahun sejak promosi terakhir" (yearsince last promotion) memiliki distribusi yang miring (skewed) dalam data. Oleh karena itu, kita akan menggunakan transformasi akar kuadrat (square root transformation).


* **Q-Q Plot:**  Plot Q-Q (Quantile-Quantile plot) adalah plot probabilitas, yaitu metode grafis untuk membandingkan dua distribusi probabilitas dengan cara memplotkan kuantilnya satu sama lain.

```python
# define a function to plot a Q-Q plot
def plot_data(data,feature):
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    data[feature].hist()
    plt.subplot(1,2,2)
    stats.probplot(data[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(data,'YearsSinceLastPromotion')
```

##**7.1 Square Root Transformation**

```python
data['square_YearsSinceLastPromotion'] = data.YearsSinceLastPromotion**(1/2)
```

```python
plot_data(data,'square_YearsSinceLastPromotion')
```

* Sekarang beberapa titik pada plot Q-Q terlihat lebih dekat dengan garis.

###**7.3.4 Principle Component Analysis**

* Principal component analysis (PCA) adalah teknik populer untuk menganalisis dataset besar yang memiliki banyak dimensi/fitur per pengamatan, meningkatkan kemudahan interpretasi data dengan tetap mempertahankan jumlah informasi yang maksimal, dan memungkinkan visualisasi data multidimensi. Secara formal, PCA adalah teknik statistik untuk mengurangi dimensionalitas suatu dataset.

* Dataset memiliki lebih dari 25 fitur, PCA dapat digunakan untuk menghilangkan noise (kebisingan) dalam data dan hanya mempertahankan fitur-fitur paling penting. Hal ini akan membantu mengatasi overfitting data dan meningkatkan kinerja model.

```python
# Getting the optimal value of PCA
from sklearn.decomposition import PCA

# Object creation
pca = PCA()

principlecomponent = pca.fit_transform(data)
plt.figure(figsize=(10,7))
sns.set_style('darkgrid')
plt.plot(np.cumsum(pca.explained_variance_ratio_),color='k')
plt.xlabel('Number of components',fontsize=20)
plt.ylabel('Variance(%)',fontsize=20) # for each componant
plt.title('Explained Variance',fontsize=20)
plt.show()
```

**Analisis:** Data tersebut memiliki total 27 fitur setelah menghapus kolom unik dan konstan. Dari analisis PCA sebelumnya, ditemukan bahwa dengan menggunakan 25 fitur, terjadi sedikit kehilangan varians. Oleh karena itu, kita akan memilih 25 fitur untuk digunakan selanjutnya.

```python
# Selecting the components
pca = PCA(n_components=25)
new_data = pca.fit_transform(data)
new_data
```
